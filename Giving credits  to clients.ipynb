### Лемматизация

по большей части нас интересует цель получения кредита поэтому разберем первым, столбец 'purpose'

```python
from pymystem3 import Mystem
m = Mystem()
df['purpose'].unique()
from collections import Counter
print(Counter(df['purpose'].unique()))
```

```python
print(df['purpose'].value_counts().head(20))
```

```python

m = Mystem()
df['purpose_lemma'] = df['purpose'].apply(m.lemmatize)
df['purpose_lemma'] = df['purpose_lemma'].replace({'жилье' : 'недвижимость'})
def purpose_category(lemmas):
    for lemma in lemmas:
        if lemma in ('недвижимость','жилье'):
            return 'Займ на покупку недвижимости' 
        if lemma in 'образование':
            return 'Займ на образование'
        if lemma in 'cвадьба':
            return 'свадьба'
        if lemma in ('автомобиль','авто'):
            return 'Займ на автомобиль'
        
#print(df['purpose_lemma'])   
df['category'] = df['purpose_lemma'].apply(purpose_category)
df['category']=df['category'].fillna('свадьба')
print(df['purpose_lemma'].head(10))
from collections import Counter
print(Counter(df['category'].unique()))
```

работающих людей разделим на категории заработка (как-никак заработок наше все)

```python
def net_worth(income):
    if income < df['total_income'].quantile(0.25):
        return 'Бедный'
    if income < df['total_income'].quantile(0.50):
        return 'Средний класс'
    if income < df['total_income'].quantile(0.75):
        return 'Аристократ'
    return 'Отец основатель' 
df['true_income'] = df['total_income'].apply(net_worth)
display(df['true_income'].value_counts())
```

### Шаг 3. Ответьте на вопросы

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

Чтобы ответить на этот вопрос сделаем сводную таблицу и разделим  наличие детей на 2 категории 0-нет.1-есть дети

```python
df['children']
def child(children):
    if children < 1:
        return 'нет детей'
    if children >= 1:
        return 'есть дети'
df['have_children']=df['children'].apply(child)
display(df['have_children'].value_counts())
```

```python
df.pivot_table(index = ['have_children'] ,columns = 'debt', values = 'dob_years', aggfunc = 'count')
debts_with_children=df.pivot_table(index = 'have_children', columns = 'debt', values = 'dob_years', aggfunc = 'count')
debts_with_children.columns = ['Нет долга', 'Есть долг']

debts_with_children['Доля должников'] = (debts_with_children['Есть долг']/
                                         (debts_with_children['Есть долг'] 
                                          + debts_with_children['Нет долга'])).map('{:.1%}'.format)
#display(debts_with_children['Доля должников'])
display(debts_with_children)
```

Как видим, доля должников среди  тех, у кого есть дети выше ,чем у бездетных)

### Вывод

Дети могут являться фактором непогашенного долга( на детей тратятся деньги и они могут являться причиной задолженностей)

- Есть ли зависимость между семейным положением и возвратом кредита в срок?

Сделаем также таблицу с долей должников!

```python
df.pivot_table(index = ['family_status'] ,columns = 'debt', values = 'dob_years', aggfunc = 'count')
status_debt=df.pivot_table(index = ['family_status'] ,columns = 'debt', values = 'dob_years', aggfunc = 'count')
status_debt.columns = ['Нет долга', 'Есть долг']
status_debt['Доля должников'] = (status_debt['Есть долг']/
                                         (status_debt['Есть долг'] 
                                          + status_debt['Нет долга'])).map('{:.1%}'.format)
#print(status_debt.head(10))
display(status_debt.sort_values(by='Доля должников',ascending=False))
```

Разделим на две группы.

```python
df['family_status']
def status_fam(marriage):
    if marriage in ('гражданский брак' and 'женат / замужем'):
        return 'пара'
    else:
        return 'один'
df['status']=df['family_status'].apply(status_fam)
display(df['status'].value_counts())
```

1)подвывод Самая низкая доля должников среди вдовцов/вдов,в то время как еще не женившиеся люди показывают отметку в 9.8 процентов

Разделим на 2 гурппы по статусу есть вторая половинка или нет:

```python
status_debt=df.pivot_table(index = ['status'] ,columns = 'debt', values = 'dob_years', aggfunc = 'count')
status_debt.columns = ['Нет долга', 'Есть долг']
status_debt['Доля должников'] = (status_debt['Есть долг']/
                                         (status_debt['Есть долг'] 
                                          + status_debt['Нет долга'])).map('{:.1%}'.format)
print(status_debt.head(10))
display(status_debt.sort_values(by='Доля должников',ascending=False))
```

### Вывод

Одинокие люди в большем количестве как должники
