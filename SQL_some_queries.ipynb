1.  Write a query to output following sales information:
	sale_date,
	client_name,
	product_name,
	total_sum - sum of daily sales
If there were no sales for the client or product during a day - exclude it from the result.
Sort by
	sale_date - ascending, total_sum – descending 


```python
sql='''select * from client;'''
```

```python
select(sql)
```

```python
sql = '''select 
  product_sale.sale_date,
  client.name,
  product.name,
  product_sale.total_sum
  from product_sale
  left join client on client.id=product_sale.id_client
  left join product on product.id =product_sale.id_product
  WHERE total_sum IS NOT NULL
  order by 
  product_sale.sale_date asc,
  product_sale.total_sum desc; '''
```

```python
select(sql)
```

3. Output sales information grouped by month and year and include following columns:
	sale_year,
	sale_month,
	client_name,
	product_name,
	total_sum - monthly sum of sales
Output only months where total_sum is less than $ 15 000

```python
sql='''select
strftime('%Y',product_sale.sale_date) AS sale_year,
strftime('%m',product_sale.sale_date) AS sale_month,
client.name as client_name,
product.name as product_name,
sum(product_sale.total_sum) as total_sum
from product_sale
  left join client on client.id=product_sale.id_client
  left join product on product.id =product_sale.id_product
where total_sum < 15000
group by
strftime('%Y',product_sale.sale_date),
strftime('%m',product_sale.sale_date),
client.name,
product.name
order by 
total_sum desc;'''
```

```python
select(sql)
```

 Write a query to output list of clients with total sales sum for 2016. Include following columns:
	client_name,
	total_sum - total yearly sum
If a client didn't purchase anything during 2016, include it with an empty total_sum

```python
sql = '''select 
  client.name as client_name,
  CASE WHEN sum(product_sale.total_sum)  IS NULL THEN 0
  ELSE sum(product_sale.total_sum) end as sum
  from client
  left join product_sale on product_sale.id_client=client.id
  where  strftime('%Y',product_sale.sale_date) ='2016' 
  group by 
  client.name
  order by sum desc;'''
```

```python
select(sql)
```

I ll make a query for ranks and then subquery

```python
sql=''' select 
  client.client_name,
  product.name,
  count(product.name),
  sum(product_sale.total_sum),
  ROW_NUMBER() OVER (
  PARTITION BY client.client_name
  ORDER BY SUM(product_sale.total_sum)  DESC
) as rating
from product_sale
join client on client.id=product_sale.id_client
join product on product.id =product_sale.id_product
where strftime('%Y',product_sale.sale_date)='2016'
group by 
client.client_name,product.name'''
```

```python
select(sql)
```

```python
sql = '''select
client_name,
name,
sum as total_sum
 from( select distinct
  client.client_name,
  product.name,
  SUM(product_sale.total_sum) sum,
  rank() OVER (
  PARTITION BY client.client_name
  ORDER BY SUM(product_sale.total_sum)  DESC
) as rating
from product_sale
join client on client.id=product_sale.id_client
join product on product.id =product_sale.id_product
where strftime('%Y',product_sale.sale_date)='2016'
group by 
client.client_name,product.name) t
where rating between 1 and 3 
order by
client_name asc,
sum  desc;'''
```

```python
select(sql)
```
